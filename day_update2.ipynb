```python
import random
from scipy.stats import norm
from scipy.stats import multivariate_normal
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
T = 10000
s = 0.5
theta_mu_orig = [11.5,19.5,15.87,5.97]
theta_std_orig = [1.63,3.46,2.31,0.69]
theta_sigma=np.diag(theta_std_orig)
thetamax = []
for i  in range(0,4):
    scale_max = theta_mu_orig[i]+5*theta_std_orig[i] 
    thetamax.append(scale_max)
thetamin =[0 for i  in range(0,4)]
theta = np.zeros((4,T+1))
np.random.seed(10)
```

```python
for i in range(0,4):
    theta[i,0] = random.uniform(thetamin[i],thetamax[i])
```

```python
def rsm_fuction(theta_start,coef):
    theta_merge = np.hstack((theta_start,theta_start**2))
    delta_tilt = np.sum(coef[:,0][1:]*theta_merge)+coef[:,0][0]
    delta_brace = np.sum(coef[:,1][1:]*theta_merge)+coef[:,1][0]
    return np.array([delta_tilt,delta_brace])
```

```python
def predict_rsm(theta_start,coef,mu,scale):
    theta_merge = np.hstack((theta_start,theta_start**2))
    predict_deflection_scale= np.sum(coef[:,0][1:]*theta_merge)+coef[:,0][0]
    predict_strain_scale= np.sum(coef[:,1][1:]*theta_merge)+coef[:,1][0]
    response_predict_scale =np.array([predict_deflection_scale,predict_strain_scale])
    response_predict = response_predict_scale*scale+mu
    return response_predict
```

```python
coef_29 = coef_df[['T29_tilt','T29_brace']].values
mu_29=mu_df[['T29_tilt','T29_brace']].values
scale_29 = scale_df[['T29_tilt','T29_brace']].values
```

```python
mu_c=np.array([0,0])
scale_c = np.array([[1,0],[0,1]])

```

```python
y_observed_27=observed_df[['T27_tilt','T27_brace']].values
coef_27 = coef_df[['T27_tilt','T27_brace']].values
mu_27 = mu_df[['T27_tilt','T27_brace']].values
scale_27 = scale_df[['T27_tilt','T27_brace']].values

```

```python
def post_dist(theta_start,y_observed,mu,scale,coef,
              theta_mu=theta_mu_orig,theta_sigma=theta_sigma,mu_c = mu_c,scale_c =scale_c):
    prior_pdf_theta = multivariate_normal.pdf(theta_start,theta_mu,theta_sigma)
    y_observed_scale = (y_observed-mu)/scale
    c_bias = y_observed_scale-rsm_fuction(theta_start,coef)
    likelihood_pdf = multivariate_normal.pdf(c_bias,mu_c,scale_c)
    return  likelihood_pdf*prior_pdf_theta
```

```python

t = 0
while t<T:
    t =t+1
    theta_start = np.random.multivariate_normal(theta[:,t-1],s*theta_sigma)
    p1 = post_dist(theta_start,y_observed_27,mu_27,scale_27,
                      coef_27)
    p2=post_dist(theta[:,t-1],y_observed_27,mu_27,scale_27,
                      coef_27)
    if p2:
        r = p1/  p2
        alpha = min(1,r)
        u = random.uniform(0,1)
        if u<alpha:
            theta[:,t] = theta_start
        else:
            theta[:,t] = theta[:,t-1]
```

```python
theta_update_27 = np.zeros(4)
theta_update_sigma_27 = theta_sigma
for  i  in range(4):
    print(np.mean(theta[i,1000:]),np.std(theta[i,1000:]))
    theta_update_27[i]=np.mean(theta[i,1000:])
    theta_update_sigma_27[i,i] = np.std(theta[i,1000:])
```

```python
for time in range(1,7):
    
    theta_tmp = np.zeros((4,T+1))
    theta_tmp[:,0]=theta_update[time-1]
    col = cols[time-1]
    y_observed_tmp=observed_df[col].values
    coef_tmp = coef_df[col].values
    mu_tmp = mu_df[col].values
    scale_tmp= scale_df[col].values
    theta_update_tmp = theta_update[time-1]
    theta_update_sigma_tmp = theta_update_sigma[time-1]
    
    t = 0
    while t<T:
        t =t+1
        theta_start = np.random.multivariate_normal(theta_tmp[:,t-1],s*theta_update_sigma_tmp)
        p1 = post_dist(theta_start,y_observed_tmp,mu_tmp,scale_tmp,
                          coef_tmp,theta_update_tmp,theta_update_sigma_tmp)
        p2=post_dist(theta_tmp[:,t-1],y_observed_tmp,mu_tmp,scale_tmp,
                           coef_tmp,theta_update_tmp,theta_update_sigma_tmp)
        if p2:
            r = p1/  p2
            alpha = min(1,r)
            u = random.uniform(0,1)
            if u<alpha:
                theta_tmp[:,t] = theta_start
            else:
                theta_tmp[:,t] = theta_tmp[:,t-1]
    theta_update_tmp = np.zeros(4)
    theta_update_sigma_tmp = theta_sigma
    print('////////////////////////////////////')
    print('this is the %d update'%time)
    for  i  in range(4):
        
        print(np.mean(theta_tmp [i,1000:]),np.std(theta_tmp[i,1000:]))
        theta_update_tmp[i]=np.mean(theta_tmp [i,1000:])
        theta_update_sigma_tmp[i,i] = np.std(theta_tmp [i,1000:])
    
    
    theta_day[time]=theta_tmp 
    theta_update[time]=theta_update_tmp
    theta_update_sigma[time] = theta_update_sigma_tmp    
    
    cols_after_tmp =cols[time:]
    predict_tilt_after_tmp=[None]*time
    predict_brace_after_tmp=[None]*time
    for col_rest in cols_after_tmp:
        coef = coef_df[col_rest].values
        mu = mu_df[col_rest].values
        scale = scale_df[col_rest].values
        predict_res = predict_rsm(theta_update_tmp,coef,mu,scale)
        predict_tilt_after_tmp.append(predict_res[0,0])
        predict_brace_after_tmp.append(predict_res[0,1])
        tilt_cols_pre =  Predict_columns[2*time]
        brace_cols_pre =  Predict_columns[2*time+1]
    Predict[tilt_cols_pre] = predict_tilt_after_tmp
    Predict[ brace_cols_pre]=predict_brace_after_tmp
```
